```python
data=pd.read_csv('MUFG train-1.csv', index_col=[0], encoding='shift-jis')

data.interpolate(inplace=True)

X = data.drop('Y', axis=1)
y = data['Y']


B=0
for i in range(20):
    X1, X2, y1, y2 = train_test_split(X, y, test_size=0.2)
    RF = RandomForestRegressor(n_estimators=500,random_state=42)
    RF.fit(X1, y1)
    Ya=RF.predict(X1)
    Yb=RF.predict(X2)
    pls=PLSRegression(n_components=6)
    pls.fit(X1,y1)
    Ya=pls.predict(X1)
    Yb=pls.predict(X2)
    print("TRAIN:",MSE(y1, Ya)**0.5, "VARIDATE",MSE(y2, Yb)**0.5)
    A = MSE(y2, Yb)**0.5
    B=A+B

C=B/20
print(C)
#print("TRAIN:",MSE(y1, Ya)**0.5, "VARIDATE",MSE(y2, Yb)**0.5)
```
